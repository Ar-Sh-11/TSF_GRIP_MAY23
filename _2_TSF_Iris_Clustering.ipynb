```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#importing the data set
import pandas as pd

irisds = pd.read_csv('https://drive.google.com/u/0/uc?id=11Iq7YvbWZbt8VXjfm06brx66b10YiwK-&export=download')
irisds.head(100)
```

```python
#perform datapreprocessing steps
irisds=irisds.dropna()
np.size(irisds)
```

```python
#summary of the dataset
irisds.describe()
```

```python
#importing Kmeans clustering algo package from sklearn
from sklearn.cluster import KMeans
```

```python
#the features are seperated from labeling
x = irisds.iloc[:, [0, 1, 2, 3]].values
x
```

```python
#computing within sum of squares or The cost function with different no. of clusters
wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', 
                    max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

```python
#plotting the cost fn(wcss) vs no of clusters to observe an elbow and by elbow method,we select the optimal number of clusters here.

plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters(K)')
plt.ylabel('WCSS(J or cost fn)') # Within cluster sum of squares
plt.show()
```

```python
#No of clusters K=3(from elbow method)
kmeans = KMeans(n_clusters = 3, init = 'k-means++', 
                    max_iter = 300, n_init = 10, random_state = 0)
y=kmeans.fit_predict(x)
y
```

```python
#visualisation of clusters -sepal length vs petal length
plt.scatter(x[y == 0, 0], x[y == 0, 1], 
            s = 100, c = 'blue', label = 'Iris-setosa')
plt.scatter(x[y == 1, 0], x[y == 1, 1], 
            s = 100, c = 'green', label = 'Iris-versicolour')
plt.scatter(x[y == 2, 0], x[y == 2, 1],
            s = 100, c = 'red', label = 'Iris-virginica')

# Plotting the centroids of the clusters
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], 
            s = 100, c = 'yellow', label = 'Centroids')
plt.xlabel('Sepal Length(in cm)')
plt.ylabel('Petal Width(in cm)')
plt.legend()
```
